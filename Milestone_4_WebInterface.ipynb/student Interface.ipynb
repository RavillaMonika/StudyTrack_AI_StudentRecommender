```python
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd
import os
from datetime import datetime

# Database file setup
if not os.path.exists('students_credentials.csv'):
    students_df = pd.DataFrame(columns=['student_id', 'name', 'email', 'password', 'registration_date'])
    students_df.to_csv('students_credentials.csv', index=False)

def registration_page():
    header = HTML("""
        <div style='text-align: center; padding: 20px; background: linear-gradient(135deg, #667eea 0%, #764ba2 100%);
                    border-radius: 10px; margin-bottom: 30px;'>
            <h1 style='color: white; margin: 0;'>üìö Study Recommender System</h1>
        </div>
    """)

    student_name = widgets.Text(
        description='Name:', placeholder='Enter your name',
        style={'description_width': '110px'}, layout=widgets.Layout(width='350px')
    )
    student_email = widgets.Text(
        description='Email ID:', placeholder='your@email.com',
        style={'description_width': '110px'}, layout=widgets.Layout(width='350px')
    )
    student_password = widgets.Password(
        description='Password:', placeholder='Enter password',
        style={'description_width': '110px'}, layout=widgets.Layout(width='350px')
    )
    save_btn = widgets.Button(
        description='Save Details',
        button_style='success',
        layout=widgets.Layout(width='180px', height='40px')
    )
    output_area = widgets.Output()

    def save_details(b):
        with output_area:
            clear_output()
            # Validation
            if not student_name.value or not student_email.value or not student_password.value:
                print("‚ùå All fields required.")
                return
            if '@' not in student_email.value:
                print("‚ùå Invalid email format.")
                return
            students_df = pd.read_csv('students_credentials.csv')
            if student_email.value in students_df['email'].values:
                print("‚ö†Ô∏è Email already registered!")
                return
            student_id = f"STU{len(students_df)+1:04d}"
            new_student = {'student_id': student_id,
                           'name': student_name.value,
                           'email': student_email.value,
                           'password': student_password.value,
                           'registration_date': datetime.now().strftime('%Y-%m-%d %H:%M:%S')}
            students_df = pd.concat([students_df, pd.DataFrame([new_student])], ignore_index=True)
            students_df.to_csv('students_credentials.csv', index=False)
            display(HTML(f"""
                <div style='background: #d4edda; border:2px solid #28a745; border-radius:7px; padding:20px; text-align:center;'>
                    <h3>‚úÖ Saved! Welcome, {student_name.value}!</h3>
                    <p>Your student ID: <b>{student_id}</b></p>
                    <p>Your details have been saved successfully.</p>
                </div>
            """))
            student_name.value = ''
            student_email.value = ''
            student_password.value = ''

    save_btn.on_click(save_details)

    display(header)
    display(student_name)
    display(student_email)
    display(student_password)
    display(HTML("<br>"))
    display(save_btn)
    display(output_area)

registration_page()
```
