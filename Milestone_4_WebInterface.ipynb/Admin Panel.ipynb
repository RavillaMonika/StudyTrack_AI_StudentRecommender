```python
# ====================================================================
# SECTION 2: ADMIN PANEL - ADMIN LOGIN
# ====================================================================

# Import required libraries
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML
import pandas as pd

def admin_panel():
    """Admin login and CRUD operations"""
    
    header = HTML("""
        <div style='text-align: center; padding: 15px; background: linear-gradient(135deg, #f093fb 0%, #f5576c 100%); 
                    border-radius: 10px; margin: 20px 0;'>
            <h2 style='color: white; margin: 0;'>üîê Admin Panel</h2>
        </div>
    """)
    
    # Admin login widgets
    admin_username = widgets.Text(
        description='Username:', 
        placeholder='admin',
        style={'description_width': '100px'}, 
        layout=widgets.Layout(width='350px')
    )
    
    admin_password = widgets.Password(
        description='Password:', 
        placeholder='Enter password',
        style={'description_width': '100px'}, 
        layout=widgets.Layout(width='350px')
    )
    
    login_btn = widgets.Button(
        description='üîì Login',
        button_style='danger',
        layout=widgets.Layout(width='150px', height='40px')
    )
    
    admin_output = widgets.Output()
    
    def admin_login(b):
        with admin_output:
            clear_output()
            
            # Check credentials
            if admin_username.value == 'admin' and admin_password.value == 'admin123':
                success_html = HTML("<div style='background: #d4edda; border:2px solid #28a745; border-radius:7px; padding:15px;'><h3 style='color: #155724;'>‚úÖ Admin Login Successful!</h3></div>")
                display(success_html)
                
                # CRUD Interface
                display(HTML("<h3 style='color: #f5576c; margin-top: 20px;'>üìä Database Management</h3>"))
                
                # View Students Button
                view_students_btn = widgets.Button(
                    description='üë• View All Students',
                    button_style='info',
                    layout=widgets.Layout(width='180px')
                )
                
                # Delete Student Input & Button
                delete_id = widgets.Text(
                    description='Student ID:',
                    placeholder='e.g., STU0001',
                    style={'description_width': '100px'},
                    layout=widgets.Layout(width='350px')
                )
                
                delete_btn = widgets.Button(
                    description='üóëÔ∏è Delete Student',
                    button_style='danger',
                    layout=widgets.Layout(width='150px')
                )
                
                crud_output = widgets.Output()
                
                def view_students(b):
                    with crud_output:
                        clear_output()
                        try:
                            students_df = pd.read_csv('students_credentials.csv')
                            if len(students_df) > 0:
                                display(HTML("<h4>All Registered Students:</h4>"))
                                display(students_df[['student_id', 'name', 'email', 'registration_date']])
                            else:
                                print("No students registered yet.")
                        except FileNotFoundError:
                            print("‚ùå Database file not found!")
                
                def delete_student(b):
                    with crud_output:
                        clear_output()
                        
                        if not delete_id.value:
                            print("‚ö†Ô∏è Please enter a Student ID!")
                            return
                        
                        try:
                            students_df = pd.read_csv('students_credentials.csv')
                            
                            if delete_id.value in students_df['student_id'].values:
                                students_df = students_df[students_df['student_id'] != delete_id.value]
                                students_df.to_csv('students_credentials.csv', index=False)
                                success_msg = HTML("<div style='background: #d4edda; border:2px solid #28a745; border-radius:7px; padding:15px;'><p style='color: #155724; margin:0;'>‚úÖ Student " + delete_id.value + " deleted successfully!</p></div>")
                                display(success_msg)
                                delete_id.value = ''
                            else:
                                error_msg = HTML("<div style='background: #f8d7da; border:2px solid #dc3545; border-radius:7px; padding:15px;'><p style='color: #721c24; margin:0;'>‚ùå Student " + delete_id.value + " not found!</p></div>")
                                display(error_msg)
                        except FileNotFoundError:
                            print("‚ùå Database file not found!")
                
                view_students_btn.on_click(view_students)
                delete_btn.on_click(delete_student)
                
                display(view_students_btn)
                display(HTML("<br>"))
                display(delete_id)
                display(delete_btn)
                display(crud_output)
                
            else:
                error_html = HTML("<div style='background: #f8d7da; border:2px solid #dc3545; border-radius:7px; padding:15px;'><p style='color: #721c24; margin:0;'>‚ùå Invalid admin credentials!</p></div>")
                display(error_html)
    
    login_btn.on_click(admin_login)
    
    display(header)
    display(admin_username)
    display(admin_password)
    display(HTML("<br>"))
    display(login_btn)
    display(admin_output)

# Call the admin panel
admin_panel()
```
